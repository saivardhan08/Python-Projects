### Data Cleaning, EDA and Statistical analysis

```python
# Importing libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading the dataset and reading the dataset

file_path = "C:/Users/vemul/OneDrive/Desktop/walmart sales/Walmart_Sales.csv"

data = pd.read_csv(file_path)

data.head(10)
```

#### Data Cleaning

```python
# removing duplicates
data.drop_duplicates(inplace=True)

data.info()
```

```python
data.describe()
```

```python
# Checking for missing values
missing_values = data.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
data[data['Date'].isna()].head(10)
```

```python
# Droping rows with missing values

data.dropna(inplace=True)

len(data)
```

```python
data.info()
```

### Future Engineering

```python
# Converting 'Date' column to datetime format

if 'Date' in data.columns:
    data['Date'] = pd.to_datetime(data['Date'],format ='%d-%m-%Y')

data.info()
```

```python
data.head(10)
```

```python
# adding relevent date-time columns for future use

data['DayOfWeek'] = data['Date'].dt.dayofweek
data['Month'] = data['Date'].dt.month
data['Year'] = data['Date'].dt.year
```

### Statistical Analysis & EDA

```python
# Correlation Matrix visualisation 

correlation_matrix = data[['Weekly_Sales','Holiday_Flag','Temperature','Fuel_Price','CPI','Unemployment','Month']].corr()
plt.figure(figsize=(10, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='Greens')
plt.title("Correlation Matrix")
plt.show()

# Another version of matrix
data[['Weekly_Sales','Holiday_Flag','Temperature','Fuel_Price','CPI','Unemployment','Month']].corr().style.background_gradient()
```

```python
#Ranking stores in descending order of weekly sales totals

data[['Store','Weekly_Sales']].groupby(['Store'],as_index=True).mean().sort_values(by='Weekly_Sales',ascending=False)
```

```python
# Top 5 Stores

top_5 = data[['Store','Weekly_Sales']].groupby(['Store'],as_index=True).mean().sort_values(by='Weekly_Sales',ascending=False).head()


print("Top 5 Stores: ") 
top_5
```

```python
# Bottom 5 Stores

Bottom_5 = data[['Store','Weekly_Sales']].groupby(['Store'],as_index=True).mean().sort_values(by='Weekly_Sales',ascending=True).head()

print("Bottom 5 Stores: ") 
Bottom_5
```

```python
# Creating Horizontal Bar for weekly sales observed over years

plt.barh(data['Year'],data['Weekly_Sales'])

plt.title('Sum of Weekly sales by year')
plt.xlabel('Weekly sales')
plt.ylabel('Year')
plt.show()
```

```python
# Visualize the data, Weekly Sales trends over time

plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='Weekly_Sales', data=data)
plt.title('Weekly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Weekly Sales')
plt.show()
```

* The sales spike during the festive season and significantly drop in the month of January

```python
# Weekly sales distribution

sns.histplot(data['Weekly_Sales'], kde=True, bins=30)
plt.title("Weekly Sales Distribution")
plt.xlabel("Weekly Sales")
plt.ylabel("Frequency")
plt.show()
```

* it appears the frequency of weekly sales hitting 0.5 Million is the highest and 3.5 Million being lowest

* and a sustained consistency of frequency is observed between ~0.7 Million and ~1.5 Million

```python
# 2.5 Sales by Store
if 'Store' in data.columns:
    store_sales = data.groupby('Store')['Weekly_Sales'].sum().reset_index()
    store_sales = store_sales.sort_values(by='Weekly_Sales', ascending=False)

    plt.figure(figsize=(12, 6))
    sns.barplot(data=store_sales, x='Store', y='Weekly_Sales', palette='viridis_r', hue='Weekly_Sales',)
    plt.title("Total Sales by Store")
    plt.xlabel("Store")
    plt.ylabel("Total Sales")
    plt.show()
```

Total Sales by each Store from 2010 - 2012 with values in Millions

```python
# 3.1 Correlation between sales and economic indicators
if 'CPI' in data.columns and 'Unemployment' in data.columns:

    sns.lmplot(data=data, x='CPI', y='Weekly_Sales', aspect=3, height=6)
    plt.title("CPI vs Weekly Sales")
    plt.show()

    sns.lmplot(data=data, x='Unemployment', y='Weekly_Sales', aspect=3, height=6)
    plt.title("Unemployment vs Weekly Sales")
    plt.show()
```

* interms of CPI a denser weekly sales or much consitent sales is seen at a lower CPI level(CPI indicates inflation)

* interestingly with unemployment rate it seems denser weekly sales are recorder at the mean point of Unemployment aroud  ~ 8% mark.

```python
# 3.2 Holiday vs Non-Holiday Sales
if 'Holiday_Flag' in data.columns:
    holiday_sales = data.groupby('Holiday_Flag')['Weekly_Sales'].mean().reset_index()

    plt.figure(figsize=(8, 6))
    sns.barplot(data=holiday_sales, x='Holiday_Flag', hue = 'Holiday_Flag' , y='Weekly_Sales', palette='coolwarm')
    plt.title("Average Weekly Sales: Holiday vs Non-Holiday")
    plt.xlabel("Is Holiday")
    plt.ylabel("Average Weekly Sales")
    plt.show()
```

1 = Holiday, 0 = Non-Holiday

Weekly sales during holiday period shows higher
